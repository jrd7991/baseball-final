```python
import numpy as np
import pandas as pd

from timeit import default_timer as timer
```

```python
eventcols = ['GAME_ID','BAT_ID','PIT_ID','EVENT_TX','EVENT_RUNS_CT']
events15 = pd.read_csv("data/events-2015.csv",usecols=eventcols)
events14 = pd.read_csv("data/events-2014.csv",usecols=eventcols)
```

```python
events15.head() # jupyter automatically prints the last line of each cell
```

```python
# date mapping example

gamecols = ['GAME_ID','GAME_DT']
games15 = pd.read_csv("data/games-2015.csv",usecols = gamecols)
games14 = pd.read_csv("data/games-2014.csv",usecols = gamecols)

games15.set_index('GAME_ID',inplace = True)
games14.set_index('GAME_ID',inplace = True)

events15['GAME_DT'] = events15['GAME_ID'].apply(lambda x: games15.get_value(x,'GAME_DT'))
events14['GAME_DT'] = events14['GAME_ID'].apply(lambda x: games14.get_value(x,'GAME_DT'))


events15.head(10)
```

```python
events = pd.concat((events15,events14))
events['DATE'] = pd.to_datetime(events['GAME_DT'],format='%Y%m%d')
events.sort_values('GAME_DT',inplace = True)

def outcomefunc(x):
        if (isinstance(x,pd.Series)): return 'NA' #to fix weird bug
        if x[0] == 'I' or x[0:2] == 'HP' or x[0] == 'W' and ((len(x) == 1) or x[1] != 'P' ):
            return 'W'
        if x[0] == 'D' and ((len(x) == 1) or x[1] != 'G'):
            return 'D'
        if x[0] == 'H' and ((len(x) == 1) or x[1] != 'P'):
            return 'H'
        if x[0] == 'S' and ((len(x) == 1) or x[1].isdigit()):
            return 'S'
        if x[0] == 'T' and ((len(x) == 1) or x[1].isdigit()):
            return 'T'
        return 'O'
    
events['D'] = events['EVENT_TX'].apply(lambda x: 1 if outcomefunc(x) == 'D' else 0)
events.head(20)
```

```python
grouped = events.groupby('BAT_ID')
```
